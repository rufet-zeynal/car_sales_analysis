```python
# Import the pandas library for data manipulation
import pandas as pd

# Import matplotlib.pyplot for plotting
import matplotlib.pyplot as plt

# Import seaborn for advanced visualizations
import seaborn as sns

# Load the CSV dataset into a pandas DataFrame
# Use raw string r'' to handle backslashes in file path
df = pd.read_csv(r'Downloads/car_sales_data.csv')

# Display the first few rows of the dataset to check it loaded correctly
df
```

```python
# Show the first 5 rows of the dataset
# Useful to quickly see how the data looks
df.head()

# Get a concise summary of the DataFrame
# Shows column names, data types, number of non-null values
df.info()

# Generate descriptive statistics for numerical columns
# Includes count, mean, std, min, 25%, 50%, 75%, max
df.describe()
```

```python
# Create a new figure (canvas) for the plot with size 7x5 inches
plt.figure(figsize=(7,5))

# Plot a histogram of the 'Price' column
# bins=30 -> divide data into 30 intervals
# kde=True -> overlay a smooth density curve on top of the histogram
sns.histplot(df['Price'], bins=30, kde=True)

# Set the title of the plot
plt.title('Distribution of Car Prices')

# Display the plot
plt.show()

```

```python
# Create a new figure (canvas) for the plot with size 7x5 inches
plt.figure(figsize=(7,5))

# Plot a histogram of the 'Mileage' column
# bins=40 -> divide the range of mileage into 40 intervals
# kde=True -> overlay a smooth density curve on top of the histogram
# color='orange' -> set the color of the bars to orange
sns.histplot(df['Mileage'], bins=40, kde=True, color='orange')

# Set the title of the plot
plt.title('Distribution of Mileage')

# Display the plot
plt.show()

```

```python
# Create a new figure (canvas) for the plot with size 6x4 inches
plt.figure(figsize=(6,4))

# Create a count plot for the 'Fuel type' column
# Each bar shows how many cars belong to each fuel type
sns.countplot(x='Fuel type', data=df)

# Set the title of the plot
plt.title('Count of Cars by Fuel Type')

# Display the plot
plt.show()

```

```python
# Create a new figure (canvas) with size 8x5 inches
plt.figure(figsize=(8,5))

# Create a boxplot to show price distribution by fuel type
# x='Fuel type' -> categories on x-axis
# y='Price' -> numerical values on y-axis
# Boxplot shows median, quartiles, and outliers for each category
sns.boxplot(x='Fuel type', y='Price', data=df)

# Set the title of the plot
plt.title('Price Distribution by Fuel Type')

# Display the plot
plt.show()

```
